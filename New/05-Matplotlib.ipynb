```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# so dont need to call plt.show()
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2

# FUNTIONAL PLOT
plt.plot(x,y,"b-")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

### Multiplot

- subplot(num of row, num of column, current)

```python
plt.subplot(1,2,1)
plt.plot(x,y);

plt.subplot(1,2,2)
plt.plot(y,x,'rx');
```

## ObjectOriented Plot

```python
# ObjecOrient

# create figure object
# - imaginary blank canvas
fig = plt.figure()

# four arg: left, bottom, width, height take value from 0-1(percent)
axes = fig.add_axes([0.1,0.1,0.4,0.4])

# plt in axes
axes.plot(x,y)
axes.set_xlabel("X axis")
axes.set_title("example");
```

```python
# playground
# - get the relationship between two axes
fig = plt.figure()

axes1 = fig.add_axes([0.0,0.0,0.8,0.8])
axes2 = fig.add_axes([0.1,0.4,0.3,0.3])

axes1.plot(x,y)
axes1.set_title("x to y")
axes2.set_title("y to x")
axes2.plot(y,x);
```

- subplots()

```python
# add subplots
# Empty canvas of 1 by 3 subplots
fig, axes = plt.subplots(nrows = 1, ncols = 3)
print type(axes) # aray - list of axes

# fix overlap
plt.tight_layout()

axes[0].plot(x,y)
axes[1].plot(y,x);
axes[2].plot(x,y,'r')
axes[2].plot(y,x,'g');
```

## Figure size, aspect ratio and DPI

- figsize is a tuple of the width and height of the figure in inches

- dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize = (8,2), dpi = 100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(figsize=(8,2),nrows=2,ncols=1, dpi = 100)

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

## Save Figure

```python
fig.savefig('picture_example.png',dpi = 200)
```

## Legend

Initial the **label="label text"** keyword argument when there are multiple plots in one ax, and then using the **legend** method without arguments to add the legend to the figure

reference: https://matplotlib.org/users/legend_guide.html#legend-location

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(y,x,label = 'x versus y')
ax.plot(x,y,label = 'y versus x');

# legend
# need to have a label first
ax.legend(loc=0)
```

# Extra Note
## Detail argument

- color: Sring likes r,b,g... or RGB Hex Code

- linewidth or lw: default = 1

- alpha: 0-1

- linestyle or ls: 'steps','-','--'..

- marker: 'o', 'x', '*', '1'..

- markersize: ...

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,alpha = 1,ls = '--',
        marker = '*', markersize = 10)
# Sring likes r,b,g... or RGB Hex Code

ax.plot(y,x,'b-x')
ax.set_xlim(-10,50)
ax.set_ylim(-10,30)
```
